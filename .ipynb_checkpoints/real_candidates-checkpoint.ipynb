```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import math as m
import matplotlib
import itertools

import dask
import dask.dataframe as dd
import dask.array as da
import dask.bag as db

import hvplot.dask
import hvplot.pandas
import timeit

%matplotlib inline

pd.set_option('display.max_columns', None)
#sn.set_palette("pastel")
```

```python
data = pd.read_csv('../real_data_candidates.csv')
```

```python
data = data[['source_id', 'q', 'q_lower', 'q_upper', 'secondary_mass', 'secondary_mass_lower', 'secondary_mass_upper', 'P', 'P_lower', 'P_upper', 'plating']]
```

```python
data
```

```python
data['Mass_1'] = (data['secondary_mass'])/(data['q'])
data['Mass_1_lower'] = (data['secondary_mass_lower'])/(data['q_upper'])
data['Mass_1_upper'] = (data['secondary_mass_upper'])/(data['q_lower'])

data['P'] = data['P']*365
data['P_lower'] = data['P_lower']*365
data['P_upper'] = data['P_upper']*365
data['logP'] = np.log10(data['P'])
data['logP_lower'] = np.log10(data['P_lower'])
data['logP_upper'] = np.log10(data['P_upper'])

data = data.rename(columns={'secondary_mass': 'Mass_BH', 'secondary_mass_upper': 'Mass_BH_upper', 'secondary_mass_lower': 'Mass_BH_lower'})

data = data[['source_id', 'Mass_BH', 'Mass_BH_lower', 'Mass_BH_upper', 'Mass_1', 'Mass_1_lower', 'Mass_1_upper', 'logP', 'logP_lower','logP_upper', 'plating']]
```

```python
data
```

```python
data.to_csv('../candidates_clean.csv')
```
