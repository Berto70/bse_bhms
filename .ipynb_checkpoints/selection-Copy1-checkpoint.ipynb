```python
outputs = dd.read_csv('./data/sevn_output_Z0.01A1L1/data_0/output_*.csv')
```

```python
outputs
```

```python
mask_NI = (outputs['BEvent'].isin([4, 5, 7, 9, 10, 11, 12, 14, 15, 17]) == False)

mask_CE = (outputs['BEvent'].isin([7, 9, 11, 12, 14, 15]))
```

```python
masked_ce = outputs[mask_CE]
```

```python
masked_CE = pd.DataFrame()
masked_NI = pd.DataFrame()

masked_CE['name'] = (masked_ce['name'].unique())
masked_CE['type'] = 'CE'
```

```python
masked_CE.head(10)
```

```python
out_masked = outputs.merge(masked_CE, on=['name'], how='left')
```

```python
mask_nce = (out_masked['type'] != 'CE')

masked_nce = out_masked[mask_nce]

masked_ni = masked_nce[mask_NI]
masked_NI['name'] = (masked_ni['name'].unique())
masked_NI['type'] = 'NI'
```

```python
out_masked_1 = out_masked.merge(masked_NI, on=['name'], how='left')
```

```python
out_masked_1.head(30)
```

```python
out_masked_1['type'] = out_masked_1['type_x'].combine_first(out_masked_1['type_y'])

# remove the cols
out_masked_1 = out_masked_1.drop(['type_x', 'type_y'], 1)

out_masked_1.head(30)
```

```python
masked_MT = pd.DataFrame()

mask_nmt = (out_masked_1['type'].isin(['IN', 'CE']) == False)

masked_mt = out_masked_1[mask_nmt]

#masked_mt = masked_nmt[mask_nmt]
masked_MT['name'] = (masked_mt['name'].unique())
masked_MT['type'] = 'MT'
```

```python
out_masked_2 = out_masked_1.merge(masked_MT, on=['name'], how='left')

out_masked_2['type'] = out_masked_2['type_x'].combine_first(out_masked_2['type_y'])

# remove the cols
out_masked_2 = out_masked_2.drop(['type_x', 'type_y'], 1)

out_masked_2.head(30)
```

```python
out_masked_2.compute().to_csv('./data/sevn_output_Z0.01A1L1/data_0.csv')
out_masked_2.compute().to_parquet('./data/sevn_output_Z0.01A1L1/data_0.parquet')
```
